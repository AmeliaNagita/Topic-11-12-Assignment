### 1. Perhatikan kolom 'Currency' dan 'Annual_Price'. Buatlah kolom baru bernama `CleanedPrice`. Kolom ini dibuat berdasarkan informasi dari 'Annual_Price' dan 'Currency', sehingga harga sewa tahunan di `CleanedPrice` memiliki mata uang yang sama, yaitu IDR (Rupiah). Gunakan Asumsi 1 USD = 15 000 IDR.

```python
df['CleanedPrice'] = df['Annual_Price']
#untuk mengisi kolom CleanedPrice dengan Annual_Price terlebih dahulu
df.loc[df['Currency'] == 'USD', 'CleanedPrice'] = df['Annual_Price'] * 15000
#diguakan untuk mengubah nilai CleanedPrice yang awalnya Annual_Price menjadi Annual_Price dikali 15000
#dikalikan 15000 karena 1 USD = 15000 IDR dan kolom Annual_Price awalnya dalam USD
#tampilkan hasil code
df
```

### 2. Analisalah kolom 'No_Rooms' dan 'CleanedPrice'. Apakah ada kejanggalan? Jika ya, nyatakanlah apa yang membuat data-data tersebut janggal. Kemudian, perbaikilah (jika hal tersebut dapat dilakukan). Jika Anda tidak memilih untuk memperbaikinya, maka apa yang sebaiknya dilakukan pada baris-baris tersebut? 

```python
df.loc[df['No_Rooms'] == 0, 'No_Rooms'] = np.nan #mengganti nilai 0 menjadi NaN, digunakan untuk mengatasi kejanggalan data
df.loc[df['No_Rooms'] > 10, 'No_Rooms'] = np.nan #mengganti nilai > 10 menjadi NaN, digunakan untuk mengatasi kejanggalan karena tidak mungkin ada rumah dengan jumlah kamar > 10


df.head()

```

Apa hasil dari analisa kolom 'No_Rooms' dan 'CleanedPrice'?

terdapat kejanggalan data pada kolom 'No_Rooms' karena terdapat nilai 0 dan > 10, yang mana tidak mungkin terjadi

kejanggalan data yang terjadi pada kolom 'No_Rooms' dan 'CleanedPrice' ,cara mengatasinya adalah dengan mengganti nilai 0 dan > 10 menjadi NaN dikarenakan tidak mungkin terjadi rumah dengan jumlah kamar 0 dan > 10 alasan tidak mengganti nilai 0 dan > 10 dengan nilai lain adalah karena tidak ada informasi lain yang dapat digunakan untuk mengganti nilai tersebut


### 3. Apartemen di Kota/Kabupaten mana yang memiliki rata-rata harga paling tinggi? Buatlah visualisasi untuk mendukung klaim Anda. 

```python
df.groupby('Region')['CleanedPrice'].mean().sort_values(ascending=False)

df.groupby('Region')['CleanedPrice'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title('Rata-rata harga apartemen berdasarkan region')
plt.xlabel('Region')
plt.ylabel('Rata-rata harga')
plt.xticks(rotation = 45)
plt.show()
```

### Analisa kolom 'Longitude' dan 'Latitude'

```python
#mengganti nilai 0 menjadi NaN, digunakan untuk mengatasi kejanggalan data
df.loc[df['Longitude'] == 0, 'Longitude'] = np.nan 
#mengganti nilai 0 menjadi NaN, digunakan untuk mengatasi kejanggalan data
df.loc[df['Latitude'] == 0, 'Latitude'] = np.nan 


df.head()
```

yang membuat data tersebut janggal adalah karena terdapat nilai 0 pada kolom Longitude dan Latitude, yang mana tidak mungkin terjadi
kejanggalan data yang terjadi pada kolom 'Longitude' dan 'Latitude' ,cara mengatasinya adalah dengan mengganti nilai 0 menjadi NaN dikarenakan tidak mungkin terjadi rumah dengan koordinat Longitude dan Latitude 0
alasan mengganti nilai NaN dengan rata-rata Longitude dan Latitude adalah karena tidak ada informasi lain yang dapat digunakan untuk mengganti nilai tersebut


### memperbaiki data yang janggal

```python
'Merubah nilai NaN menjadi rata-rata Longitude'

df.loc[df['Longitude'].isnull(), 'Longitude'] = df['Longitude'].mean() 

'Merubah nilai NaN menjadi rata-rata Latitude'

df.loc[df['Latitude'].isnull(), 'Latitude'] = df['Latitude'].mean() 

'Menampilkan hasil perbaikan data yang janggal berupa tabel'
#menampilkan data yang memiliki nilai NaN pada kolom Longitude
df.loc[df['Longitude'].isnull()] 
#menampilkan data yang memiliki nilai NaN pada kolom Latitude
df.loc[df['Latitude'].isnull()] 
df.sample(5)

```

Apa hasil dari analisa kolom 'Longitude' dan 'Latitude' ?

kejanggalan data yang terjadi pada kolom 'Longitude' dan 'Latitude' ,cara mengatasinya adalah dengan mengganti nilai 0 menjadi NaN dikarenakan tidak mungkin terjadi rumah dengan koordinat Longitude dan Latitude 0
alasan tidak mengganti nilai 0 dengan nilai lain adalah karena tidak ada informasi lain yang dapat digunakan untuk mengganti nilai tersebut
